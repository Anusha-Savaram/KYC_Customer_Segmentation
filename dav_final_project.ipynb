```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
ss.fit(df1)
ssdf1 = pd.DataFrame(ss.transform(df1),columns= df1.columns )
ssdf1.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(ssdf1)
pca_df = pd.DataFrame(pca.transform(ssdf1), columns = ['col1', 'col2', 'col3'])
```

```python
pca.explained_variance_ratio_
```

```python

from sklearn.cluster import KMeans
```

```python
k = range(1,20)
inertias = []

for i in k:
  model = KMeans(n_clusters=i)
  model.fit(pca_df)
  inertias.append(model.inertia_)

plt.figure(figsize=(15,10))
plt.plot(k, inertias, '-o')
plt.xlabel('K')
plt.ylabel('ineria')
plt.xticks(k)
plt.title("Choosing k value",fontsize=21,weight="bold")
plt.show()
```

```python
k_means = KMeans(n_clusters = 4, random_state = 50)
y_pred = k_means.fit_predict(pca_df)
pca_df['Cluster'] = y_pred
```

```python
df['Cluster'] = y_pred
```

```python
Sum_of_products_all_age_group=df.groupby(by=["Age"]).sum()[['Wines','Fruits','Meat','Fish','Sweet','Gold']].reset_index()
```

```python
Sum_of_products_all_age_group.head()
```

```python
def return_highest(cols):
    Wines=cols[0]
    Fruits=cols[1]
    Meat=cols[2]
    Fish=cols[3]
    Sweet=cols[4]
    Gold=cols[5]
    if (Wines>=Fruits) and (Wines>=Meat) and (Wines>=Fish) and  (Wines>=Sweet) and (Wines>=Gold):
        return 'Wines'
    elif (Fruits>=Wines) and (Fruits>=Meat) and (Fruits>=Fish) and  (Fruits>=Sweet) and (Fruits>=Gold):
        return 'Fruits'
    elif (Meat>=Wines) and (Meat>=Fruits) and (Meat>=Fish) and  (Meat>=Sweet) and (Meat>=Gold):
        return 'Meat'
    elif (Fish>=Wines) and (Fish>=Fruits) and (Fish>=Meat) and  (Fish>=Sweet) and (Fish>=Gold):
        return 'Fish'
    elif (Sweet>=Wines) and (Sweet>=Fruits) and (Sweet>=Meat) and  (Sweet>=Fish) and (Fish>=Gold):
        return 'Sweet'
    elif (Gold>=Wines) and (Gold>=Fruits) and (Gold>=Meat) and  (Gold>=Fish) and (Gold>=Sweet):
        return 'Gold'
```

```python
Sum_of_products_all_age_group['product_consumed'] = Sum_of_products_all_age_group[['Wines', 'Fruits', 'Meat', 'Fish', 'Sweet', 'Gold']].apply(return_highest,axis=1)
```

```python
Middle_age=Sum_of_products_all_age_group[(Sum_of_products_all_age_group['Age']>=36) & (Sum_of_products_all_age_group['Age']<=45)]
```

```python
Upper_Middle_age=Sum_of_products_all_age_group[(Sum_of_products_all_age_group['Age']>=46) & (Sum_of_products_all_age_group['Age']<=59)]
```

```python
Senior_Adult=Sum_of_products_all_age_group[Sum_of_products_all_age_group['Age']>=60]
```
